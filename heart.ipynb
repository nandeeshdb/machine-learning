```python
import  pandas as pd
path="C:\\Users\\Win10\\Desktop\\internship\\ip\\Logistic\\Heart Failure Prediction\\heart.csv"
data=pd.read_csv(path)
print(data)
print(data.info())
```

```python
data
```

```python
import sklearn

from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
le_ChestPainType=LabelEncoder()
le_RestingECG=LabelEncoder()
le_ExerciseAngina=LabelEncoder()
le_ST_Slope=LabelEncoder()
        
data['Sex_n']=le_Sex.fit_transform(data['Sex'])
data['ChestPainType_n']=le_ChestPainType.fit_transform(data['ChestPainType'])
data['RestingECG_n']=le_RestingECG.fit_transform(data['RestingECG'])
data['ExerciseAngina_n']=le_ExerciseAngina.fit_transform(data['ExerciseAngina'])
data['ST_Slope_n']=le_ST_Slope.fit_transform(data['ST_Slope'])





```

```python
input=data.drop(['HeartDisease','Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],'columns')

print(input)
```

```python
output=data.drop(['Age','Cholesterol','FastingBS','RestingBP','MaxHR','Oldpeak','Sex_n','ChestPainType_n','RestingECG_n','ExerciseAngina_n','ST_Slope_n','Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],'columns')
print(output)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,output,test_size=0.8)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
res=model.predict([[40,1,1,140,289,0 ,1,172,0,0.0,2]])
print(res)
```
