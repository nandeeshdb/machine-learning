```python
import pandas as pd
path="C:\\Users\\Win10\\Desktop\\internship\\ip\\Logistic\\ignore_Cirrhosis Prediction\\cirrhosis.csv"
data=pd.read_csv(path)
print(data)
```

```python
a=data.Cholesterol.median()
print(a)# for finding median value
temp=data.Cholesterol.fillna(a)#uses to fill empty data in temp variable
print(temp)

data.Cholesterol=data.Cholesterol.fillna(a)

```

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
le_Status=LabelEncoder()
le_Drug=LabelEncoder()
le_Sex=LabelEncoder()
le_Ascites=LabelEncoder()
le_Hepatomegaly=LabelEncoder()
le_Spiders=LabelEncoder()
le_Edema=LabelEncoder()
data['Status_n']=le_Status.fit_transform(data['Status'])
data['Drug_n']=le_Drug.fit_transform(data['Drug'])
data['Sex_n']=le_Sex.fit_transform(data['Sex'])
data['Ascites_n']=le_Ascites.fit_transform(data['Ascites'])
data['Hepatomegaly_n']=le_Hepatomegaly.fit_transform(data['Hepatomegaly'])
data['Spiders_n']=le_Spiders.fit_transform(data['Spiders'])
data['Edema_n']=le_Edema.fit_transform(data['Edema'])
data.drop(['Status','Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema'],'columns')
print(data)
```

```python
input=data.drop(['ID','Status','Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema','Stage'],'columns')
print(input)
```

```python
output=data.drop(['ID','Status','Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema','N_Days','Age','Bilirubin','Cholesterol','Albumin','Copper','Alk_Phos','SGOT','Tryglicerides','Platelets','Prothrombin','Status_n','Drug_n','Sex_n','Ascites_n','Hepatomegaly_n','Spiders_n','Edema_n'],'columns')

print(output)
```

```python
import sklearn 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,output,train_size=0.8)# 0.8 is eqaul to 80% for traing
print(x_train)
print(x_test)
```

```python
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()#model installing
model.fit(x_train,y_train)
y_pred=model.predict(x_test)# input is testing
#y_pred-predict output from the model


print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,y_test)
print(cm)
```
