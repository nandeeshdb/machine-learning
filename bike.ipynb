```python
import pandas as pd
path="C:\\Users\\Win10\\Desktop\\internship\\ip\\Linear_MultipleRegression\\Bike sharing rental prediction hourly\\hour.csv"
data=pd.read_csv(path)
print(data)
```

```python
input=data.drop(['instant','dteday','cnt'],'columns')
print(input)
```

```python
output=data.drop(['instant','dteday','season','yr','mnth','hr','holiday','weekday','workingday','weathersit','temp','atemp','hum','windspeed','casual','registered'],'columns')
print(output)
```

```python
import sklearn#sk leaern is the machine learing algorithm
from sklearn import linear_model
model=linear_model.LinearRegression()#installing the model
model.fit(input,output)#traing the model

m=model.coef_
b=model.intercept_
print(b)


print(m)
res=model.predict([[1,0,1,0,0,6,0,1,0.24,0.2879,0.81,0,12,1123]])
print(res)
```
